```python
class ContaBancaria:
    
    def __init__(self, saldo_inicial):
        self.saldo = saldo_inicial
        self.extrato = []
        self.verifica_saques = []
        self.lista_usuarios = []
        
    def deposito(self):
        
        valor = float(input("Digite o valor: "))

        if valor > 0:
            self.saldo += valor
            self.extrato.append(f"Depósito: R$ {valor:.2f}")

        else:
           print('Por favor, digite um valor válido.')

    def saque(self):
        
        QUANTIDADE_SAQUES = 3
        LIMITE_SAQUE = 500
    
        valor = float(input("Digite o valor a sacar: "))

        if valor > 0:

            if valor <= LIMITE_SAQUE:
            
                if len(self.verifica_saques) < QUANTIDADE_SAQUES:

                    if self.saldo >= valor:
                        self.saldo -= valor
                        self.extrato.append(f"Saque: R$ {valor:.2f}")
                        self.verifica_saques.append(1)

                    else:
                        print("Operação falhou! Você não tem saldo suficiente.")

                else:
                    print("Limite de saques diários atingido.")

            else:
                print("Por favor, digite um valor abaixo do limite R$ 500,00.")       

    def exibir_extrato(self):

        print("Extrato:")

        for operacao in self.extrato:
            print(operacao)
        print(f'\nSaldo atual R$ {self.saldo:.2f}')
```

```python
def menu():

    print(
    """

    Opções:
    1. Depositar
    2. Sacar
    3. Exibir Extrato
    0. Sair      
    """
    )


# Função para ler a opção digitada pelo usuário
def ler_opcao():

    while True:

        try:

            opcao = int(input("Digite a opção desejada: "))
            if (opcao >= 0) and (opcao <= 3):
                return opcao
            else:
                print("Opção inválida. Digite novamente.")
        except:
            print("Entrada inválida. Digite novamente.")
```

```python
def main():

    conta = ContaBancaria(saldo_inicial = 0)

    while True:

        menu()
        opcao = ler_opcao()

        if opcao == 1:
            conta.deposito()

        elif opcao == 2:
        
            conta.saque()
        
        elif opcao == 3:
            conta.exibir_extrato()
        
        elif opcao == 0:
            print("Saindo do programa.")
            break

if __name__ == "__main__":
    main()
```
