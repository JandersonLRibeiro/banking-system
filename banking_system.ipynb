```python
from datetime import datetime
```

```python
class ContaBancaria:
    
    def __init__(self):
        
        self.lista_clientes = [] # Para armazenar as contas criadas no banco
        
        self.cpf =[] # Para armazenar os CPF's cadastrados
        
        self.conta = [] # Para gerar os números das contas
        
    def cadastrar_usuario(self):
        
        
        cpf = input("\nDigite o CPF (apenas números): ")     
                    
        if cpf in self.cpf:
                
            print("\nJá existe uma conta com esse CPF!")
            
        else:
            
            self.cpf.append(cpf)
            
            nome = input("\nDigite o nome completo: ")
            
            data_nascimento = input("\nDigite a data de nascimento (dd-mm-aaaa): ")
            
            endereco = input("\nDigite o endereço (logradouro, nro - bairro - cidade/sigla estado): ")
            
            usuario = {       
            "cpf" : cpf,
            "nome" : nome,
            "data_nascimento" : data_nascimento,
            "endereco": endereco
            }
            
            self.lista_clientes.append(usuario)
            
            print("\nUsuário criado com sucesso!")          
                
    def cadastrar_conta_corrente(self):
            
        cpf = input("\nDigite o CPF (apenas números): ")
            
        if cpf not in self.cpf:
                
            print("Este CPF não está cadastrado, para executar essa operação cadastre o usuário")
        
        else:
                
            AGENCIA = "0001"
                    
            self.conta.append(1)
                
            for cliente in self.lista_clientes:
                    
                if cliente["cpf"] == cpf:
                        
                    cliente["agencia"] = AGENCIA
                    
                    cliente["numero_conta"] = len(self.conta)
                    
                    cliente["data_hora"] = datetime.now().strftime("%d-%m-%Y %H:%M")
                    
                    cliente["saldo"] = []
                    
                    cliente["extrato"] = []
                    
                    cliente["quantidade_saques"] = []
                                                                     
                    print("\n=" * 40)
                    
                    print("              CONTA CORRENTE")
                    
                    print("=" * 40)
                    
                    nome = cliente["nome"]
                    
                    agencia = cliente["agencia"]
                    
                    conta = cliente["numero_conta"]
                    
                    data_hora = cliente["data_hora"]
            
                    print(f"Nome do Titular: {nome}")
            
                    print(f"Agência: {agencia}")
            
                    print(f"Número da Conta: {conta}")
            
                    print(f"Data e Hora de Criação: {data_hora}")
            
                    print("-" * 40)     
            
    def deposito(self):
        
        cpf = input("Digite o CPF (apenas números): ")
            
        if cpf in self.cpf:
                
            numero_conta = int(input("Digite o número da conta: "))
                
            for cliente in self.lista_clientes:    
                    
                if (cliente["cpf"] == cpf) and (cliente["numero_conta"] == numero_conta):                    
                    
                    valor = float(input("Digite o valor: "))
                        
                    if valor > 0:
                        
                        cliente["saldo"].append(valor)
                        
                        data_deposito = datetime.now().strftime("%d-%m-%Y %H:%M")
                            
                        cliente["extrato"].append(f"{data_deposito} - Depósito: R$ {valor:.2f}")
                            
                    else:
                            
                        print("Por favor, digite um valor válido")
        else:
                
            print("CPF não encontrado. Para fazer esse operação cadastre-se e abra uma conta.")
        
    def saque(self):
        
        QUANTIDADE_SAQUES = 3
        
        LIMITE_SAQUE = 500
        
        saldo_acumulado = []
        
        cpf = input("Digite o CPF (apenas números): ")
        
        if cpf in self.cpf:
            
            numero_conta = int(input("Digite o número da conta: "))
            
            for cliente in self.lista_clientes:
                
                saldo = cliente["saldo"]
                
                saldo_acumulado.append(saldo)
                
                if (cliente["cpf"] == cpf) and (cliente["numero_conta"] == numero_conta):
                    
                    if len(cliente["quantidade_saques"]) < QUANTIDADE_SAQUES:
                            
                        valor = float(input("Digite o valor: "))
                        
                        if valor > 0:
                            
                            if valor < LIMITE_SAQUE:
                                
                                if sum(saldo_acumulado[numero_conta-1]) >= valor:
                                    
                                    cliente["quantidade_saques"].append(1)
                                    
                                    cliente["saldo"].append(-valor)
                        
                                    data_saque = datetime.now().strftime("%d-%m-%Y %H:%M")
                            
                                    cliente["extrato"].append(f"{data_saque} - Saque: R$ {valor:.2f}")
                                
                                else:
                                    
                                    print("Operação falhou! Você não tem saldo suficiente.")        
                            else:
                            
                                print("Por favor, digite um valor abaixo do limite R$ 500,00.")   
                        else:
                                                                       
                            print("Por favor, digite um valor válido.")
                            
                    else:
                    
                        print("Limite de saques diários atingido.")  
        
    def exibir_extrato(self):
        
        contas = []
        
        cpfs = []
        
        nomes = []
        
        operacoes = []
        
        saldo_acumulado =  []
                        
        for cliente in self.lista_clientes:
            
            numero_conta = cliente["numero_conta"]
            
            cpf = cliente["cpf"]
            
            nome = cliente["nome"]
            
            operacao = cliente["extrato"]
            
            saldo = cliente["saldo"]
            
            contas.append(numero_conta)
            
            cpfs.append(cpf)
            
            nomes.append(nome)
            
            operacoes.append(operacao)
            
            saldo_acumulado.append(saldo)
            
        conta_corrente = int(input("Digite o número da Conta Corrente: "))  
        
        for conta in contas:
                
            if conta == conta_corrente:
                                
                print("=" * 40)
            
                print("              CONTA CORRENTE")
        
                print("=" * 40)
        
                print(f"Nome do Titular: {nomes[conta_corrente-1]}")
                
                cpf = cpfs[conta_corrente-1]
            
                print(f"CPF: {cpf[:3]}.{cpf[3:6]}.{cpf[6:9]}-{cpf[9:]}")
            
                print(f"Número da Conta: {contas[conta_corrente-1]}")
            
                print("-" * 40)
            
                print("           TRANSIÇÕES RECENTES")
            
                print("-" * 40)
            
                for operacao in operacoes[conta_corrente-1]:
                    
                    print(operacao)
                    
                print(f'\nSaldo atual R$ {sum(saldo_acumulado[conta_corrente-1]):.2f}')       
```

```python
def menu():

    print(
    """

    Opções:
    1. Cadastrar Usuário
    2. Cadastrar Conta Corrente
    3. Depositar
    4. Sacar
    5. Exibir Extrato
    0. Sair      
    """
    )


# Função para ler a opção digitada pelo usuário
def ler_opcao():

    while True:

        try:

            opcao = int(input("Digite a opção desejada: "))
            if (opcao >= 0) and (opcao <= 5):
                return opcao
            else:
                print("Opção inválida. Digite novamente.")
        except:
            print("Entrada inválida. Digite novamente.")
```

```python
def main():

    conta = ContaBancaria()

    while True:

        menu()
        opcao = ler_opcao()
        
                    
        if opcao == 1:
            conta.cadastrar_usuario()
            
        elif opcao == 2:
            conta.cadastrar_conta_corrente()

        elif opcao == 3:
            conta.deposito()

        elif opcao == 4:
        
            conta.saque()
        
        elif opcao == 5:
            conta.exibir_extrato()

        elif opcao == 0:
            print("Saindo do programa.")
            break

if __name__ == "__main__":
    main()
```
